```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/세종교육/iris.csv", sep=',')
df
```

```python
# 데이터 정규
df['x1'] = (df['x1'] - df['x1'].min()) / (df['x1'].max() - df['x1'].min())
df['x2'] = (df['x2'] - df['x2'].min()) / (df['x2'].max() - df['x2'].min())
df['x3'] = (df['x3'] - df['x3'].min()) / (df['x3'].max() - df['x3'].min())
df['x4'] = (df['x4'] - df['x4'].min()) / (df['x4'].max() - df['x4'].min())
```
