```python
import math

def solution(numer1, denom1, numer2, denom2):
    lcm = int(abs(denom1 * denom2 // math.gcd(denom1, denom2))) #최대공약수
    return [int((numer1*(lcm//denom1))+(numer2*(lcm//denom2))),lcm]
```

```python
print(solution(1,6,1,8))
```

```python
print(solution(5,5,5,5))
```

```python
print(solution(3,4,5,6))
```
